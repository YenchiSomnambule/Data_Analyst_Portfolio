### Date Background

The data to be used for analysis is the California Housing Prices dataset available on Kaggle (https://www.kaggle.com/datasets/camnugent/california-housing-prices/data).

This dataset includes features such as longitude, latitude, housing median age, total rooms, total bedrooms, population, households, median income, median house value, and ocean proximity. These features are essential to the data mining goals of predicting housing values in California, understanding the influences on these prices, and identifying optimal investment opportunities in the real estate market. The quality of the dataset is adequate for analysis, as it provides a comprehensive list of variables and optimaized size.

### Explaination of each data feature:

1. Longitude: Indicates the longitudinal position of a house, with higher values corresponding to a greater distance towards the west.
2. Latitude: Indicates the latitudinal position of a house, with higher values indicating a greater distance towards the north.
3. Housing Median Age: Represents the median age of houses within a block, where a lower value suggests newer constructions.
4. Total Rooms: Indicates the overall count of rooms within a block.
5. Total Bedrooms: Indicates the total count of bedrooms within a block.
6. Population: Indicates the total population residing within a block.
7. Households: Refers to the total count of households within a block, each representing a group of individuals living within a single housing unit.
8. Median Income: Reflects the median income for households within a block, measured in tens of thousands of US Dollars.
9. Median House Value: Indicates the median value of houses within a block, measured in US Dollars.
10. Ocean Proximity: Describes the location of the house in relation to the ocean or sea.

### Data Visualization

#### Box Plot

```python
# Boxplot for "medianHouseValue", "medianIncome", and "housingMedianAge"
plt.figure(figsize=(18, 6))

# Plotting "medianHouseValue"
plt.subplot(1, 3, 1)
sns.boxplot(y=df['median_house_value'])
plt.title('Boxplot of Median House Value')

# Plotting "medianIncome"
plt.subplot(1, 3, 2)
sns.boxplot(y=df['median_income'])
plt.title('Boxplot of Median Income')

# Plotting "housingMedianAge"
plt.subplot(1, 3, 3)
sns.boxplot(y=df['housing_median_age'])
plt.title('Boxplot of Housing Median Age')

plt.tight_layout()
plt.show()
```

```python
# Plotting the boxplots
plt.figure(figsize=(18, 6))

# Boxplot by Ocean Proximity for Median House Value
plt.subplot(1, 2, 1)
sns.boxplot(x='ocean_proximity', y='median_house_value', data=df)
plt.title('Median House Value by Ocean Proximity')
plt.xticks(rotation=45)

# Boxplot of Households
plt.subplot(1, 2, 2)
sns.boxplot(y='households', data=df)
plt.title('Households in each block')

plt.tight_layout()
plt.show()
```

#### Distribution Plot

```python
# Setting up the figure for combined plots
plt.figure(figsize=(18, 6))

# First plot for 'median_house_value'
plt.subplot(1, 3, 1)
sns.histplot(df, x='median_house_value', kde=True)
plt.title('Distribution of Median House Value')

# Second plot for 'median_income'
plt.subplot(1, 3, 2)
sns.histplot(df, x='median_income', kde=True)
plt.title('Distribution of Median Income')

# Third plot for 'housing_median_age'
plt.subplot(1, 3, 3)
sns.histplot(df, x='housing_median_age', kde=True)
plt.title('Distribution of Housing Median Age')

plt.tight_layout()
plt.show()
```

#### Correlation

```python
# Selecting the columns of interest
columns = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Calculating the correlation matrix
correlation_matrix = df[columns].corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Housing Attributes')
plt.show()
```

```python
# Define the columns for the pair plot
columns_of_interest = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Generate the pair plot
sns.pairplot(df[columns_of_interest])
plt.show()
```

#### Additional relevant Plots

```python

# Scatter Plot with Geographical Data for House Value
plt.figure(figsize=(10, 8))
plt.scatter(df['longitude'], df['latitude'], c=df['median_house_value'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of House Values')
plt.show()

# Scatter Plot with Geographical Data for Income
plt.figure(figsize=(10, 8))
plt.scatter(df['longitude'], df['latitude'], c=df['median_income'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Median Income')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Median Income')
plt.show()

# Bar Plot for Average House Value by Ocean Proximity
avg_value_by_proximity = df.groupby('ocean_proximity')['median_house_value'].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_value_by_proximity.plot(kind='bar')
plt.xlabel('Ocean Proximity')
plt.ylabel('Average Median House Value')
plt.title('Average Median House Value by Ocean Proximity')
plt.xticks(rotation=45)
plt.show()
```

According to this observation, an experimental predictive formula can be constructed as: House_Value = Population_Income * PI_weight + Location_Type * LT_weight + constant.
In the above formula, Location_Type would be classified as "Inland", "<1HR Ocrean", "Near Ocean", "Near Bay", and "Island".
